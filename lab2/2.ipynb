```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def match_histograms(input_image, reference_image):
    """Match the histogram of the input image to the reference image."""
    matched = np.zeros_like(input_image)
    for i in range(input_image.shape[-1]):  # Process each channel separately
        input_hist, bins = np.histogram(input_image[..., i].ravel(), 256, [0, 256])
        reference_hist, bins = np.histogram(reference_image[..., i].ravel(), 256, [0, 256])
        
        # Compute cumulative distribution function (CDF)
        input_cdf = np.cumsum(input_hist) / input_hist.sum()
        reference_cdf = np.cumsum(reference_hist) / reference_hist.sum()
        
        # Create a mapping from input image to reference image
        mapping = np.interp(input_cdf, reference_cdf, np.arange(256))
        
        # Apply the mapping
        matched[..., i] = cv2.LUT(input_image[..., i], mapping.astype(np.uint8))
    return matched

def display_images(input_img, reference_img, matched_img):
    """Display the input, reference, and matched images along with histograms."""
    images = [input_img, reference_img, matched_img]
    titles = ["Input Image", "Reference Image", "Histogram Matched Image"]
    
    plt.figure(figsize=(18, 6))
    for i, img in enumerate(images):
        plt.subplot(2, 3, i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(titles[i])
        plt.axis('off')
        
        plt.subplot(2, 3, i + 4)
        for j in range(3):  # RGB channels
            hist = cv2.calcHist([img], [j], None, [256], [0, 256])
            plt.plot(hist, label=f'Channel {j}')
        plt.title(f"Histogram - {titles[i]}")
        plt.legend()
    plt.show()

def main():
    # Load input and reference images
    input_path = "/home/student/Pictures/Screenshots/Screenshot from 2024-12-04 11-50-25.png"  # Path to input image
    reference_path = "/home/student/Pictures/Screenshots/Screenshot from 2024-12-03 11-45-00.png"  # Path to reference image
    
    input_img = cv2.imread(input_path)
    reference_img = cv2.imread(reference_path)
    
    if input_img is None or reference_img is None:
        print("Error: Could not load images. Please check file paths.")
        return
    
    # Match histograms
    matched_img = match_histograms(input_img, reference_img)
    
    # Display results
    display_images(input_img, reference_img, matched_img)
    
    # Save matched image
    cv2.imwrite("matched_image.jpg", matched_img)
    print("Histogram matched image saved as 'matched_image.jpg'")

if __name__ == "__main__":
    main()
```
