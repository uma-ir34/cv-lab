```python
import cv2
from matplotlib import pyplot as plt
# Read the image
image_path = '/home/student/Pictures/Screenshots/Screenshot from 2024-12-03 12-23-34.png'  # Replace with your image path
image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Specify the coordinates of the pixel (row, column)
row = 100  # Row index of the pixel
col = 150  # Column index of the pixel

# Extract the BGR values of the pixel
bgr_pixel = image[row, col]

# Convert BGR to RGB
rgb_pixel = (bgr_pixel[2], bgr_pixel[1], bgr_pixel[0])

# Display the RGB values
print(f"The RGB values of the pixel at ({row}, {col}) are: {rgb_pixel}")
```
