```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def harris_corner_detection(image_path):
    # Read the image
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply Harris corner detection
    gray = np.float32(gray)
    dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
    
    # Dilate the corners to enhance visualization
    dst = cv2.dilate(dst, None)
    
    # Threshold for marking the corners
    img[dst > 0.01 * dst.max()] = [0, 0, 255]
    
    # Display the results
    plt.figure(figsize=(10, 7))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Harris Corner Detection")
    plt.axis('off')
    plt.show()
def fast_corner_detection(image_path):
    # Read the image
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply FAST corner detection
    fast = cv2.FastFeatureDetector_create()
    keypoints = fast.detect(gray, None)
    
    # Draw the keypoints
    img_keypoints = cv2.drawKeypoints(img, keypoints, None, color=(255, 0, 0))
    
    # Display the results
    plt.figure(figsize=(10, 7))
    plt.imshow(cv2.cvtColor(img_keypoints, cv2.COLOR_BGR2RGB))
    plt.title("FAST Corner Detection")
    plt.axis('off')
    plt.show()
# Paths to images
image1 = '/home/student/Pictures/Screenshots/Screenshot from 2024-12-07 11-42-05.png'
image2 = '/home/student/Pictures/Screenshots/Screenshot from 2024-12-05 09-42-07.png'

# Apply Harris Corner Detection
harris_corner_detection(image1)
harris_corner_detection(image2)

# Apply FAST Corner Detection
fast_corner_detection(image1)
fast_corner_detection(image2)
```
